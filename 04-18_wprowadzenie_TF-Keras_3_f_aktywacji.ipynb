```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import math

np.set_printoptions(precision=6)
```

```python
def max_relu(x):
  return max(x, 0.0)
```

```python
for x in np.arange(-3.0,3.0, 0.5):
  print(max_relu(x))
```

```python
x = np.arange(-3.,3.,0.5)
```

```python
x
```

```python
plt.scatter(x=x, y=[max_relu(i) for i in x])
```

```python
data = sorted(data)
```

```python
def sigm(x):
  return 1 / (1+np.exp(-x))
```

```python
df = pd.DataFrame({'data':data, 'sigmoid':[sigm(x) for x in data]})
```

```python
plt.scatter(df['data'], df['sigmoid'])
```

```python
px.line(data_frame=df, x='data', y='sigmoid')
```

```python
def tahn(x):
  return (np.exp(x)-np.exp(-x)) / (np.exp(x)+np.exp(-x))
```

```python
data = 2 * np.random.randn(100)
data = sorted(data)
data
```

```python
tanh_data = [tahn(x) for x in sorted(data)]
tanh_data
```

```python
df = pd.DataFrame({'data': data, 'tanh_data': tanh_data})
df.head()
```

```python
px.line(df, x='data', y='tanh_data', width=700, height=400, title='Tanh Function')
```

```python
def softmax(x):
    e_x = np.exp(x)
    denominator = np.sum(e_x, axis=1)
    denominator = denominator[:, np.newaxis]
    return e_x / denominator
```

```python

data = np.random.randn(4, 5)
data
```

```python
result = softmax(data)
result
```
